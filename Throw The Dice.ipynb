### Two persons roll a dice in order and get points depending on the number that appears. If each of them gets a 6, they have the right to throw the dice once more. In addition, if each of them attains 10,000 games, the game ends. Get each individual's average score by simulating in Python.

```python
# Import maduls 
import random
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
```

```python
result_1=[]
result_2=[]

Number_of_players=2

for i in range(1,10001):
    for j in range(1, Number_of_players+1):
        if j==1:
            result1=np.random.randint(1,7) 
            result_1.append(result1)
            if result1==6:
                result1=np.random.randint(1,7) 
                result_1.append(result1)
        elif j==2:
            result2=np.random.randint(1,7) 
            result_2.append(result2)
            if result2==6:
                result2=np.random.randint(1,7) 
                result_2.append(result2)
        while len(result_1)==10000 | len(result_2)==10000:
            break

Final_result_1 = "The results of the first person's dice roll are as follows : {} " .format(result_1)
Final_result_2 = "The results of the second person's dice roll are as follows : {} " .format(result_2)

print (Final_result_1)
print("The average count obtained by the first player in the dice is.", stat.mean (result_1))

print (Final_result_2)
print("The average count obtained by the Second player in the dice is.", stat.mean(result_2))
```

```python
Number_of_elements_1 = []
Number_of_elements_2 = []

for i in range (1,7):
    Number_of_elements_1.append(result_1.count(i))
    Number_of_elements_2.append(result_2.count(i))
    
print(Number_of_elements_1) 
print(Number_of_elements_2)


   
```
